```python
import numpy as np
```

```python
dt = np.dtype([('time', np.unicode_, 16), ('ubication',np.unicode_, 16)])
prueba = []
a = np.arange(5)
b = np.arange(1,29)
prueba.clear()
print("g20-2")
for j in b:
    for i in a:
        datos = np.loadtxt("results_LocGlo_"+str(j)+"/results/"+str(j)+"/g20-2/result"+str(i)+"_Hora.txt", delimiter=" " ,dtype=dt)
        prueba.append(str(datos)[2:10])
    print(prueba)
    prueba.clear()
```

```python
with open('tiempoTotal.txt','w') as file:
    for i in a:
        for j in b:
            datos = np.loadtxt("results_LocGlo_"+str(j)+"/results/"+str(j)+"/g20-2/result"+str(i)+"_Hora.txt", delimiter=" " ,dtype=dt)
            c = str(datos)[2:10]
            file.write(c+' ')
        file.write('\n')

prueba.clear()
print("g20-2")
for j in b:
    for i in a:
        datos = np.loadtxt("results_LocGlo_"+str(j)+"/results/"+str(j)+"/g20-2/result"+str(i)+"_Hora.txt", delimiter=" " ,dtype=dt)
        prueba.append(str(datos)[2:10])
    print(prueba)
    prueba.clear()
print("g20-4")
for j in b:
    for i in a:
        datos = np.loadtxt("results_LocGlo_"+str(j)+"/results/"+str(j)+"/g20-4/result"+str(i)+"_Hora.txt", delimiter=" " ,dtype=dt)
        prueba.append(str(datos)[2:10])
    print(prueba)
    prueba.clear()
print("gvar-2")
for j in b:
    for i in a:
        datos = np.loadtxt("results_LocGlo_"+str(j)+"/results/"+str(j)+"/gvar-2/result"+str(i)+"_Hora.txt", delimiter=" " ,dtype=dt)
        prueba.append(str(datos)[2:10])
    print(prueba)
    prueba.clear()
```
